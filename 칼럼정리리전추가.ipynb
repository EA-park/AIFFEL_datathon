```python
import pandas as pd

report2015 = pd.read_csv('/Users/frodo/Downloads/archive/2015.csv')
report2016 = pd.read_csv('/Users/frodo/Downloads/archive/2016.csv')
report2017 = pd.read_csv('/Users/frodo/Downloads/archive/2017.csv')
report2018 = pd.read_csv('/Users/frodo/Downloads/archive/2018.csv')
report2019 = pd.read_csv('/Users/frodo/Downloads/archive/2019.csv')

# 다섯 개 파일 모두 변수에 할당

print(set(report2015['Country']) - set(report2016['Country']))
print(set(report2016['Country']) - set(report2015['Country']))

region = {}
for i, j in zip(report2015['Country'], report2015['Region']):
    region[i] = j
    
for i, j in zip(report2016['Country'], report2016['Region']):
    region[i] = j
    
region
```

```python
report2015.columns = ['Country', 'Region', 'Rank', 'Score', 'Standard Error', 'GDP', 'Social Support',
                      'Healthy Life Expectancy', 'Freedom', 'Corruption', 'Generosity', 'Dystopia Residual']
report2016.columns = ['Country', 'Region', 'Rank', 'Score', 'LCI', 'UCI', 'GDP', 'Social Support',
                      'Healthy Life Expectancy', 'Freedom', 'Corruption', 'Generosity', 'Dystopia Residual']
report2017.columns = ['Country', 'Rank', 'Score', 'UCI', 'LCI', 'GDP', 'Social Support',
                      'Healthy Life Expectancy', 'Freedom', 'Generosity', 'Corruption', 'Dystopia Residual']
report2018.columns = [ 'Rank', 'Country', 'Score', 'GDP', 'Social Support', 'Healthy Life Expectancy',
                      'Freedom', 'Generosity', 'Corruption']
report2019.columns = [ 'Rank', 'Country', 'Score', 'GDP', 'Social Support', 'Healthy Life Expectancy',
                      'Freedom', 'Generosity', 'Corruption']

# 연도 별 파일들의 칼럼 이름을 통일

report2015.drop(['Standard Error', 'Dystopia Residual'], axis = 1, inplace = True)
report2016.drop(['LCI', 'UCI', 'Dystopia Residual'], axis = 1, inplace = True)
report2017.drop(['UCI', 'LCI', 'Dystopia Residual'], axis = 1, inplace = True)

# 사용하지 않을 칼럼들을 삭제
```
